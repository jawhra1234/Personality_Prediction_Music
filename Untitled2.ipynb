```python
# Complete 7-Stage EDA for Music Dataset
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import skew, kurtosis
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# ==========================================
# STAGE 1: DATA COLLECTION & LOADING
# ==========================================
print("="*60)
print("STAGE 1: DATA COLLECTION & LOADING")
print("="*60)

# Load the dataset
df = pd.read_csv('music_data.csv')

print(f"\n✓ Dataset loaded successfully!")
print(f"✓ Shape: {df.shape[0]} rows × {df.shape[1]} columns")

# ==========================================
# STAGE 2: DATA STRUCTURE UNDERSTANDING
# ==========================================
print("\n" + "="*60)
print("STAGE 2: DATA STRUCTURE UNDERSTANDING")
print("="*60)

print("\n2.1 First Few Records:")
print(df.head())

print("\n2.2 Dataset Info:")
print(df.info())

print("\n2.3 Column Names:")
print(df.columns.tolist())

print("\n2.4 Data Types Summary:")
print(df.dtypes.value_counts())

print("\n2.5 Memory Usage:")
print(f"Total memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# ==========================================
# STAGE 3: DATA CLEANING & QUALITY CHECK
# ==========================================
print("\n" + "="*60)
print("STAGE 3: DATA CLEANING & QUALITY CHECK")
print("="*60)

print("\n3.1 Missing Values Analysis:")
missing = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df) * 100).round(2)
})
missing = missing[missing['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
print(missing if len(missing) > 0 else "✓ No missing values found!")

print("\n3.2 Duplicate Rows:")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

print("\n3.3 Unique Values per Column:")
unique_counts = df.nunique().sort_values(ascending=False)
print(unique_counts)

print("\n3.4 Data Quality Issues:")
# Check for potential issues
issues = []

# Check for negative durations
if 'duration_ms' in df.columns and (df['duration_ms'] < 0).any():
    issues.append("- Negative duration values found")

# Check for year range
if 'year' in df.columns:
    year_range = f"Year range: {df['year'].min()} to {df['year'].max()}"
    print(year_range)
    if df['year'].min() < 1900 or df['year'].max() > 2025:
        issues.append("- Unusual year values detected")

if issues:
    print("Issues detected:")
    for issue in issues:
        print(issue)
else:
    print("✓ No major data quality issues detected")

# ==========================================
# STAGE 4: UNIVARIATE ANALYSIS
# ==========================================
print("\n" + "="*60)
print("STAGE 4: UNIVARIATE ANALYSIS")
print("="*60)

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"\n4.1 Numerical Features ({len(numerical_cols)}):")
print(numerical_cols)

print(f"\n4.2 Categorical Features ({len(categorical_cols)}):")
print(categorical_cols)

print("\n4.3 Descriptive Statistics for Numerical Features:")
print(df[numerical_cols].describe().round(3))

print("\n4.4 Statistical Properties:")
stats_df = pd.DataFrame({
    'Mean': df[numerical_cols].mean(),
    'Median': df[numerical_cols].median(),
    'Std': df[numerical_cols].std(),
    'Skewness': df[numerical_cols].skew(),
    'Kurtosis': df[numerical_cols].kurtosis()
}).round(3)
print(stats_df)

# Visualizations for Univariate Analysis
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols[:12]):
    axes[idx].hist(df[col].dropna(), bins=50, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'Distribution of {col}', fontsize=10, fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('music_univariate_distributions.png', dpi=300, bbox_inches='tight')
print("\n✓ Univariate distribution plots saved as 'music_univariate_distributions.png'")
plt.show()

# Box plots for outlier detection
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols[:12]):
    axes[idx].boxplot(df[col].dropna(), vert=True)
    axes[idx].set_title(f'Boxplot of {col}', fontsize=10, fontweight='bold')
    axes[idx].set_ylabel(col)
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('music_univariate_boxplots.png', dpi=300, bbox_inches='tight')
print("✓ Boxplot analysis saved as 'music_univariate_boxplots.png'")
plt.show()

# Categorical analysis
print("\n4.5 Categorical Features Analysis:")
for col in categorical_cols[:5]:  # Limit to first 5 categorical columns
    print(f"\n{col}:")
    print(df[col].value_counts().head(10))

# ==========================================
# STAGE 5: BIVARIATE ANALYSIS
# ==========================================
print("\n" + "="*60)
print("STAGE 5: BIVARIATE ANALYSIS")
print("="*60)

print("\n5.1 Correlation Matrix:")
correlation_matrix = df[numerical_cols].corr()
print(correlation_matrix.round(3))

# Correlation heatmap
plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap of Musical Features', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('music_correlation_heatmap.png', dpi=300, bbox_inches='tight')
print("\n✓ Correlation heatmap saved as 'music_correlation_heatmap.png'")
plt.show()

print("\n5.2 Strong Correlations (|r| > 0.5):")
strong_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.5:
            strong_corr.append({
                'Feature 1': correlation_matrix.columns[i],
                'Feature 2': correlation_matrix.columns[j],
                'Correlation': correlation_matrix.iloc[i, j]
            })

if strong_corr:
    strong_corr_df = pd.DataFrame(strong_corr).sort_values('Correlation', 
                                                            ascending=False, 
                                                            key=abs)
    print(strong_corr_df)
else:
    print("No strong correlations found")

# Scatter plots for key relationships
if 'energy' in df.columns and 'loudness' in df.columns:
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.ravel()
    
    relationships = [
        ('energy', 'loudness'),
        ('danceability', 'valence'),
        ('acousticness', 'energy'),
        ('tempo', 'energy'),
        ('popularity', 'danceability'),
        ('speechiness', 'acousticness')
    ]
    
    for idx, (x, y) in enumerate(relationships):
        if x in df.columns and y in df.columns:
            axes[idx].scatter(df[x], df[y], alpha=0.5, s=20)
            axes[idx].set_xlabel(x, fontsize=10)
            axes[idx].set_ylabel(y, fontsize=10)
            axes[idx].set_title(f'{x} vs {y}', fontsize=11, fontweight='bold')
            
            # Add regression line
            z = np.polyfit(df[x].dropna(), df[y].dropna(), 1)
            p = np.poly1d(z)
            axes[idx].plot(df[x], p(df[x]), "r--", alpha=0.8, linewidth=2)
            axes[idx].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('music_bivariate_relationships.png', dpi=300, bbox_inches='tight')
    print("\n✓ Bivariate scatter plots saved as 'music_bivariate_relationships.png'")
    plt.show()

# ==========================================
# STAGE 6: MULTIVARIATE ANALYSIS
# ==========================================
print("\n" + "="*60)
print("STAGE 6: MULTIVARIATE ANALYSIS")
print("="*60)

print("\n6.1 Feature Interactions:")

# Pair plot for selected features (subset to avoid overplotting)
key_features = ['valence', 'energy', 'danceability', 'acousticness', 'loudness']
key_features = [f for f in key_features if f in df.columns][:5]

if len(key_features) >= 3:
    sample_df = df[key_features].sample(min(1000, len(df)), random_state=42)
    pairplot = sns.pairplot(sample_df, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 30})
    pairplot.fig.suptitle('Pairwise Relationships (Sample)', y=1.02, fontsize=16)
    plt.savefig('music_multivariate_pairplot.png', dpi=300, bbox_inches='tight')
    print("✓ Pairplot saved as 'music_multivariate_pairplot.png'")
    plt.show()

# Group analysis
if 'year' in df.columns and 'popularity' in df.columns:
    print("\n6.2 Temporal Analysis:")
    df['decade'] = (df['year'] // 10) * 10
    decade_stats = df.groupby('decade')[numerical_cols].mean()
    print(decade_stats.round(3))
    
    # Plot trend over time
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.ravel()
    
    trend_features = ['energy', 'danceability', 'valence', 'acousticness']
    for idx, feature in enumerate(trend_features):
        if feature in df.columns:
            decade_mean = df.groupby('decade')[feature].mean()
            axes[idx].plot(decade_mean.index, decade_mean.values, 
                          marker='o', linewidth=2, markersize=8)
            axes[idx].set_xlabel('Decade', fontsize=11)
            axes[idx].set_ylabel(f'Average {feature}', fontsize=11)
            axes[idx].set_title(f'{feature.capitalize()} Trend Over Time', 
                              fontsize=12, fontweight='bold')
            axes[idx].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('music_temporal_trends.png', dpi=300, bbox_inches='tight')
    print("\n✓ Temporal trends saved as 'music_temporal_trends.png'")
    plt.show()

# ==========================================
# STAGE 7: INSIGHTS & RECOMMENDATIONS
# ==========================================
print("\n" + "="*60)
print("STAGE 7: INSIGHTS & RECOMMENDATIONS")
print("="*60)

print("\n7.1 KEY FINDINGS:")
print("-" * 60)

# Finding 1: Dataset Overview
print(f"\n📊 Dataset Overview:")
print(f"   • Total tracks analyzed: {len(df):,}")
print(f"   • Features available: {len(df.columns)}")
print(f"   • Time span: {df['year'].min()} - {df['year'].max()}" if 'year' in df.columns else "")

# Finding 2: Feature Distributions
print(f"\n📈 Feature Characteristics:")
for col in ['danceability', 'energy', 'valence', 'acousticness']:
    if col in df.columns:
        print(f"   • {col.capitalize()}: mean={df[col].mean():.3f}, std={df[col].std():.3f}")

# Finding 3: Outliers
print(f"\n⚠️  Outlier Detection:")
for col in numerical_cols[:5]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()
    if outliers > 0:
        print(f"   • {col}: {outliers} outliers ({outliers/len(df)*100:.2f}%)")

# Finding 4: Correlations
print(f"\n🔗 Notable Correlations:")
if strong_corr:
    for item in strong_corr_df.head(3).to_dict('records'):
        print(f"   • {item['Feature 1']} ↔ {item['Feature 2']}: {item['Correlation']:.3f}")

print("\n7.2 RECOMMENDATIONS FOR ML:")
print("-" * 60)

recommendations = [
    "1. Feature Engineering:",
    "   • Create interaction terms for highly correlated features",
    "   • Consider decade-based features for temporal patterns",
    "   • Normalize features with different scales (e.g., duration_ms vs popularity)",
    "",
    "2. Data Preprocessing:",
    "   • Handle outliers in loudness and tempo using robust scaling",
    "   • Apply log transformation to skewed features (duration_ms, tempo)",
    "   • Consider one-hot encoding for categorical variables (artists, mode, key)",
    "",
    "3. Feature Selection:",
    "   • Remove highly correlated features (|r| > 0.9) to reduce multicollinearity",
    "   • Use PCA if dimensionality reduction is needed",
    "   • Keep acousticness, energy, valence as they show distinct patterns",
    "",
    "4. Model Considerations:",
    "   • For popularity prediction: Use regression models (Random Forest, XGBoost)",
    "   • For genre classification: Use ensemble methods with these audio features",
    "   • Consider time-based train-test split to avoid data leakage",
    "",
    "5. Validation Strategy:",
    "   • Use stratified k-fold for classification tasks",
    "   • Implement time-series split if temporal dependency exists",
    "   • Monitor for overfitting with features like instrumentalness (high skew)"
]

for rec in recommendations:
    print(rec)

print("\n7.3 DATA QUALITY SUMMARY:")
print("-" * 60)
print(f"✓ Missing values: {df.isnull().sum().sum()} ({df.isnull().sum().sum()/df.size*100:.2f}%)")
print(f"✓ Duplicate rows: {duplicates}")
print(f"✓ Numerical features: {len(numerical_cols)}")
print(f"✓ Categorical features: {len(categorical_cols)}")
print(f"✓ Ready for ML: {'Yes' if df.isnull().sum().sum() == 0 else 'Requires cleaning'}")

print("\n" + "="*60)
print("EDA COMPLETE! All visualizations saved.")
print("="*60)
```

```python
# =====================================================
# PHASE 2: PERSONALITY-BASED MUSIC RECOMMENDATION
# =====================================================

import pickle
import pandas as pd
import numpy as np

# --- Load Trained Models ---
traits = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']
models = {}

for t in traits:
    with open(f"{t}_final_model.pkl", "rb") as f:
        models[t] = pickle.load(f)

print("✅ All models loaded successfully!")

# --- Load Dataset ---
df = pd.read_csv("music_data.csv")

# Keep only relevant columns for scoring
features = ['danceability', 'energy', 'valence', 'acousticness', 'instrumentalness', 'tempo']
df = df[['name', 'artists'] + features].dropna()

# --- Simulated Personality Input ---
# (You can replace this with your model predictions later)
predicted_traits = {
    'Extraversion': 3.376,
    'Agreeableness': 1.357,
    'Conscientiousness': 3.323,
    'Neuroticism': 2.13,
    'Openness': 3.131
}

print("\n🎯 Predicted Personality Traits:")
for trait, val in predicted_traits.items():
    print(f"{trait}: {val:.3f}")

# --- Scoring Function ---
def personality_to_song_score(personality, song):
    score = (
        personality['Extraversion'] * (song['energy'] + song['danceability']) / 2 +
        personality['Agreeableness'] * (song['valence'] + song['acousticness']) / 2 +
        personality['Conscientiousness'] * (1 - abs(song['tempo'] - 120) / 120) +
        (5 - personality['Neuroticism']) * song['valence'] +
        personality['Openness'] * song['instrumentalness']
    )
    return score

# --- Compute Scores for All Songs ---
df['score'] = df.apply(lambda row: personality_to_song_score(predicted_traits, row), axis=1)

# --- Sort by Score ---
df_sorted = df.sort_values(by='score', ascending=False).reset_index(drop=True)

# --- Select Random 5 Songs from Top 50 ---
top_50 = df_sorted.head(50)
recommended_songs = top_50.sample(5, random_state=None)  # randomness adds natural variety

# --- Display Recommendations ---
print("\n🎵 Top 5 Recommended Songs (Randomized from Top 50):")
display(recommended_songs[['name', 'artists', 'score']])
```
